```python
import matplotlib.pyplot as plt
import numpy as np

rows = 2
cols = 3
fig, axs = plt.subplots(rows, cols, figsize=(10, 10))
for i in range(rows):
    for j in range(cols):
        axs[i, j].imshow(np.random.rand(256, 256, 3))
        axs[i, j].set_title('Image {}'.format(i*cols+j))
        axs[i, j].axis('off')
fig.tight_layout()
plt.show()
```

```python
a = np.random.rand(256, 256, 3)
b = np.dstack([a for _ in range(3)])     # stack the images along depth channel
print('shape of a:', a.shape)
print('shape of b:', b.shape)
```
